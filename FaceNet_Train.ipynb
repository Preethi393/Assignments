```python
import numpy as np
import os
import matplotlib.pyplot as plt
from deepface.commons import functions
from deepface.basemodels import Facenet
import faiss
import glob
```

```python
model = Facenet.loadModel()
```

```python
representations = []
for face in glob.glob('/home/ubuntu/Desktop/Face_Recognition_DS_team/Train/*/*'):
    img = functions.preprocess_face(img=face, target_size=(160, 160),enforce_detection=False)
    embedding = model.predict(img)[0,:]
    representation = []
    representation.append(face)
    representation.append(embedding)
    representations.append(representation)
    
```

```python
embedding.shape
```

```python
embeddings = []
for i in range(0, len(representations)):
    embedding = representations[i][1]
    embeddings.append(embedding)
    
embeddings = np.array(embeddings, dtype='f')
```

```python
embeddings.shape
```

```python
dimensions = 128 #FaceNet output is 128 dimensional vector
 
metric = 'euclidean' #euclidean, cosine

if metric == 'euclidean':
    index = faiss.IndexFlatL2(dimensions)
elif metric == 'cosine':
        index = faiss.IndexFlatIP(dimensions)
        faiss.normalize_L2(embeddings)
```

```python
index.add(embeddings)
```

```python
#save
faiss.write_index(index,"vect.index")
 
#restore
index = faiss.read_index("vect.index")
```

```python
def nearest_neighbors(path):
    target_img = functions.preprocess_face(img=path,
                                       target_size=(160, 160), enforce_detection=False)
    target_representation = model.predict(target_img)[0,:]
    target_representation = np.array(target_representation,dtype='f')
    target_representation = np.expand_dims(target_representation,axis = 0)
    
    k = 4
    distances, neighbors = index.search(target_representation,k)
    distances = distances[0]
    neighbors = neighbors[0]
    return distances, neighbors
```
