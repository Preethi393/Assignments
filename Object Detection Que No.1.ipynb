```python
annotation_file = glob ('/home/ubuntu/Desktop/animal detection.v7-semi-fixed-labels-640x640.darknet/train/*.txt')
annotation_file
```

```python
def read_annotation_file(annotation_file):
    # Read the annotation file and return the bounding box coordinates and class labels
    # Assumes the annotation file has the format: [xmin, ymin, xmax, ymax, class]
    df = pd.read_csv(annotation_file, header=None, delimiter=' ')
    df.columns = ['class','xmin', 'ymin', 'xmax', 'ymax']
    df['image_path'] = annotation_file.replace('.txt', '.jpg')
    return df

# Iterate over the annotation files and read them
for file in annotation_file:
    annotations = read_annotation_file(file)
    # Do something with the annotations DataFrame
    print(annotations.head())
```

#### Class Count:

```python
# Update the code to use the correct paths and DataFrame
df = pd.DataFrame()  # Initialize an empty DataFrame

for file in annotation_file:
    annotations = read_annotation_file(file)
    df = df.append(annotations, ignore_index=True)
```

```python
df.head()
```

```python
def get_class_count(df):
    # Count the occurrences of each class in the dataset
    class_count = df['class'].value_counts()
    return class_count

class_count = get_class_count(df)
print(class_count)
```

```python
def plot_class_count(class_count):
    # Plot the class count using a bar chart
    plt.bar(class_count.index, class_count.values)
    plt.xlabel('Class')
    plt.ylabel('Count')
    plt.title('Class Count')
    plt.show()
```

```python
class_count = get_class_count(df)
plot_class_count(class_count)
```

#### Bounding boxes per image:

```python
def get_bounding_boxes_per_image(df):
    # Count the number of bounding boxes per image
    bounding_boxes_per_image = df['image_path'].value_counts()
    return bounding_boxes_per_image
```

```python
def plot_bounding_boxes_per_image(bounding_boxes_per_image):
    # Plot the bounding boxes per image
    bounding_boxes_per_image.plot(kind='hist', bins=20)
    plt.xlabel('Number of Bounding Boxes')
    plt.ylabel('Count')
    plt.title('Bounding Boxes per Image')
    plt.show()
```

```python
bounding_boxes_per_image = get_bounding_boxes_per_image(df)
plot_bounding_boxes_per_image(bounding_boxes_per_image)
```

#### Bounding boxes per each class:

```python
def get_bounding_boxes_per_class(df):
    # Count the number of bounding boxes per class
    bounding_boxes_per_class = df['class'].value_counts()
    return bounding_boxes_per_class
```

```python
def plot_bounding_boxes_per_class(bounding_boxes_per_class):
    # Plot the bounding boxes per class
    bounding_boxes_per_class.plot(kind='bar')
    plt.xlabel('Class')
    plt.ylabel('Count')
    plt.title('Bounding Boxes per Class')
    plt.show()
```

```python
bounding_boxes_per_class = get_bounding_boxes_per_class(df)
plot_bounding_boxes_per_image(bounding_boxes_per_class)
```

#### Distribution of Bounding box size:

```python
def calculate_bounding_box_sizes(df):
    # Calculate the sizes of bounding boxes (width, height)
    df['width'] = df['xmax'] - df['xmin']
    df['height'] = df['ymax'] - df['ymin']
    return df
```

```python
def plot_bounding_box_size_distribution(df):
    # Plot the distribution of bounding box sizes
    plt.hist(df['width'], bins='auto', alpha=0.5, label='Width')
    plt.hist(df['height'], bins='auto', alpha=0.5, label='Height')
    plt.xlabel('Size')
    plt.ylabel('Count')
    plt.title('Bounding Box Size Distribution')
    plt.legend()
    plt.show()
```

```python
df = calculate_bounding_box_sizes(df)
plot_bounding_box_size_distribution(df)
```
