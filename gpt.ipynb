```python
import os
import json
files=[]
path='/content/drive/MyDrive/chat_gpt_files/'
for file in os.listdir(path):
    file_text=os.path.join(path,file)
    with open(file_text)as f:
        data = json.load(f)
        files.append(data)
```

```python
print(data)
```

```python
from langchain.text_splitter import CharacterTextSplitter, NLTKTextSplitter, SpacyTextSplitter
text_splitter = CharacterTextSplitter(        
    separator = "\n\n",
    chunk_size = 1600,
    chunk_overlap  = 200,
    length_function = len,
)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
docs = []
metadatas = []
for doc in files:
    text = doc["Content"]
    source = doc["Source"]
    splits = text_splitter.split_text(text)
    for splited_text in splits:
        docs.append(splited_text)
        metadatas.append({"source":source})
```

```python
#  embeddings
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
# vectore stores
from langchain.vectorstores import FAISS
```

```python
import faiss
import pickle
import torch
store = FAISS.from_texts(docs[:100],embeddings, metadatas=metadatas[:100])
faiss.write_index(store.index, "docs.index")
store.index = None
with open("vectors_stores.pkl", "wb") as f:
    pickle.dump(store, f)
    
```

```python
index = faiss.read_index("docs.index")
with open("vectors_stores.pkl", "rb") as f:
    store = pickle.load(f)
store.index = index
```
