```python
# Create dummy variables using the 'get_dummies' command
dummy = pd.get_dummies(df[['Lead Origin', 'Lead Source', 'Do Not Email', 'Last Activity',
                              'What is your current occupation','A free copy of Mastering The Interview', 
                              'Last Notable Activity']], drop_first=True)

# Add the results to the master dataframe
df = pd.concat([df, dummy], axis=1)
```

```python
# Creating dummy variable separately for the variable 'Specialization' since it has the level 'Select' which is useless so we drop that level by specifying it explicitly

dummy_spl = pd.get_dummies(df['Specialization'], prefix = 'Specialization')
dummy_spl = dummy_spl.drop(['Specialization_Select'], 1)
df = pd.concat([df, dummy_spl], axis = 1)
```

```python
# Drop the variables for which the dummy variables have been created

df = df.drop(['Lead Origin', 'Lead Source', 'Do Not Email', 'Last Activity',
                   'Specialization', 'What is your current occupation',
                   'A free copy of Mastering The Interview', 'Last Notable Activity'], 1)
```

```python
df
```

```python
# Put all the feature variables in X

X = df.drop(['Converted'], 1)
X.head()
```

```python
# Put the target variable in y

y = df['Converted']

y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the dataset into 70% train and 30% test

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)

print('Shape of training data:',(X_train.shape))
print('Shape of testing data:',(X_test.shape))
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
X_train[['TotalVisits','Total Time Spent on Website','Page Views Per Visit']] = scaler.fit_transform(X_train[['TotalVisits','Total Time Spent on Website','Page Views Per Visit']])

X_test[['TotalVisits','Total Time Spent on Website','Page Views Per Visit']] = scaler.fit_transform(X_test[['TotalVisits','Total Time Spent on Website','Page Views Per Visit']])
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier()
# Import RFE model for feature selection
from sklearn.feature_selection import RFE

rfe= RFE(rf_model,n_features_to_select=10)

rfe.fit(X_train,y_train)
```

```python
pd.DataFrame(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
top_10_col=X_train.columns[rfe.support_]
top_10_col
```

```python
X_train = X_train[top_10_col]
X_train
```

```python
X_test = X_test[top_10_col]
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,plot_roc_curve
```

```python
print(classification_report(y_test,pred))
```

```python
confusion_matrix(y_test,pred)
```

```python
accuracy_score(y_test,pred)
```
